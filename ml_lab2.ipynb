```python
import pandas as pd

# Load dataset
df = pd.read_csv("Car_sales.csv")

# Function to find min and max values of any column
def find_min_max(df, column_name):
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in DataFrame.")

    min_value = df[column_name].min()
    max_value = df[column_name].max()

    # Find the rows with the min and max values
    min_row = df[df[column_name] == min_value]
    max_row = df[df[column_name] == max_value]

    return min_value, max_value, min_row, max_row

# Now call the function to find min and max for Width
width_min, width_max, width_min_row, width_max_row = find_min_max(df, "Width")

# Print results
print("\nWidth - Min:", width_min, "\n", width_min_row)
print("\nWidth - Max:", width_max, "\n", width_max_row)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('Car_sales.csv')
print("Column names in the dataset:")
print(df.columns)
sales_column = [col for col in df.columns if 'sales' in col.lower()]
if not sales_column:
    raise KeyError("No column related to 'sales' found in the dataset.Please check the column names.")
    sales_column_name = sales_column[0]
    print(f"Using column '{sales_column_name}' as the sales column.")
    y = df[sales_column_name]
    X = df.drop(columns=[sales_column_name])
    categorical_columns = X.select_dtypes(include=['object','category']).columns
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])
        label_encoders[col] = le
        print("\nShape of X (input):", X.shape)
        print("Shape of y (output):", y.shape)
        print("\nInput (X) with encoded categorical variables:")
        print(X.head())
        print("\nOutput (y):")
        print(y.head())
        for col in categorical_columns:
            print(f"\nMapping for {col}:")
            print(dict(zip(label_encoders[col].classes_,
            label_encoders[col].transform(label_encoders[col].classes_))))
```

```python
import pandas as pd

def find_min_max(df, column_name):
    # Check if column exists
    if column_name not in df.columns:
        return None, f"Error: Column '{column_name}' not found in the DataFrame."

    # Convert column to numeric (set non-numeric values to NaN)
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

    # Compute min and max
    min_value = df[column_name].min()
    max_value = df[column_name].max()

    # Handle cases where all values are NaN
    if pd.isna(min_value) or pd.isna(max_value):
        return None, f"Error: Column '{column_name}' contains only invalid or missing values."

    return min_value, max_value

# Load dataset
df = pd.read_csv("Car_sales.csv")

# Define column name
column_name = "Sales in thousands"

# Call the function
min_value, max_value = find_min_max(df, column_name)

# Print results
if min_value is None:
    print(max_value)
else:
    print(f"Min Value of '{column_name}': {min_value}")
    print(f"Max Value of '{column_name}': {max_value}")
```

```python
# 11. Sort data by Price and find the most & least expensive cars
if "Price" in df.columns:
    df["Price"] = pd.to_numeric(df["Price"], errors="coerce")  # Ensure numeric type
    df.dropna(subset=["Price"], inplace=True)  # Drop rows with missing prices

    if not df.empty:
        df.sort_values(by="Price", ascending=True, inplace=True)
        print("\n Least Expensive Car:\n", df.iloc[0])
        print("\n Most Expensive Car:\n", df.iloc[-1])
    else:
        print("\n No valid price data available after cleaning.")
else:
    print("\n 'Price' column not found! Available columns:", df.columns.tolist())

```
